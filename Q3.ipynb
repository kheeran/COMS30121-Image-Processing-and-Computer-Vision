```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

def imshow(image):
    #OpenCV stores images in BGR so we have to convert to RGB to display it using matplotlib
    imagergb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(imagergb)


#edge detection
ddx = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
# ddx = np.ones((3,3))
ddy = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])



location = str("images/dart") + str(6) + str(".jpg")
imagecol = cv2.imread(location)
imagegray = cv2.cvtColor(imagecol, cv2.COLOR_BGR2GRAY)
(M,N) = imagegray.shape

#init img
image = np.zeros((M+2,N+2))

for i in range (1,M+1):
    for j in range (1,N+1):
        image[i,j] = imagegray[i-1,j-1]

dxedgeimg = np.zeros((M,N))

for i in range (1,M+1):
    for j in range (1,N+1):
        sum = 0
        for a in range (0,2):
            for b in range (0,2):
                #image[i-1+a,j-1+b] to start from the top left coner of the 3x3 grid
                sum += image[i-1+a,j-1+b]*ddx[2-a,2-b] 
        
        dxedgeimg[i-1,j-1] = sum

dyedgeimg = np.zeros((M,N))

for i in range (1,M+1):
    for j in range (1,N+1):
        sum = 0
        for a in range (0,2):
            for b in range (0,2):
                #image[i-1+a,j-1+b] to start from the top left coner of the 3x3 grid
                sum += image[i-1+a,j-1+b]*ddy[2-a,2-b] 
        dyedgeimg[i-1,j-1] = sum
        
plt.imshow(dyedgeimg, cmap='gray')
```
